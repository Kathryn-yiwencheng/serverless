 ## Section 3 Application for ETL pipeline from Section 1 using serverless AWS services with each component and its function within the pipeline.

### Raw data 
- Different raw data sources. 
- File type includes: csv, json, sftp and so on. 
- The raw data after transformation can use feature engineering for machine learning models. 
 
 
### Amazon Kinesis
- Extracting data real-time from the raw data
- Kinesis Client Library (KCL) makes it more similar to Kafka. 
- It is managed by AWS which has higher stability without in-house maintenance. 
 
 
### Amazon S3
- Amazon Simple Storage Service.
- Retain raw data in a S3 bucket. 
- S3 is a data storage for all climate data before aggregation / data cleaning. 
- Another S3 bucket stored transformed data
 
 
### Amazon Elastic MapReduce (EMR)
- One use case is to use EMR launch Spark jobs as a ETL tool 
    - Spark provides distributed job execution.
    - Distributed collections are provided for unstructured data (RDDs)
    - Filters, Map operations, reduce operations may be done on RDD 
    - Distributed tables for structured data (dataframes)
    - SQL-like dataframe operations may be performed on dataframes
    - Dataframes may be registered as “tables” for standard SQL queries. 
- Elastic MapReduce is Amazon’s framework for managing Hadoop and Spark.
- Provides a mechanism to launch a YARN cluster at the desired scale.
- Provides a packaged Spark distribution as well as other tools in the Hadoop ecosystem. 
- Allows scaling up with larger instances. 
- Allows scaling out with more nodes. 
- Can utilize either S3 or HDFS for scalable storage.  
  
### Amazon Athena
- Data Lake store data without fixed data schema. 
- AWS Athena is a perfect tool because it is easy to scale. 
- Why should we choose AWS Athena instead of AWS Redshift / Snowflake ? 
    - Schema is defined on data stored in S3. Schema may be stored in a hive meta store or AWS glue. Glue provide a crawler to discover tables and infer schema. 
    - Athena provides a fully-managed query service that takes SQL and generates tables as results. 
    - Allows querying of uncompressed and compressed CSV files, JSON objects, as well as Parquet and ORC columnar formats. 
Business Intelligence tools as well as other applications can connect via JDBC or ODBC.
